```python
# reconizing activities based on sensory data
import pandas as pd
import os

DATA_PATH = 'datasets/Activity Recognition from Single Chest-Mounted Accelerometer'
def load_data_for_person(person_id):
    csv_path = os.path.join(DATA_PATH, (person_id +'.csv'))
    X= pd.read_csv(csv_path, header=None)
    X.rename(columns={0:"SeqNum", 1:"X", 2: "Y", 3:"Z", 4:"Label"}, inplace=True)
    return X
```
